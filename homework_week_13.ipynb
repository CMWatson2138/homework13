### The next step is to handle large outliers.  I checked for outliers using the describe method.  

```python
#Looking at what the data is comprised of.  Surpressed scientific notation with lambda
clean_bot_df.describe().apply(lambda s: s.apply('{0:.5f}'.format))
```

```python
mod_bot_df=clean_bot_df.copy()
# IQR
Q1 = mod_bot_df.quantile(0.25)
Q3 = mod_bot_df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
mod_bot_df_out = mod_bot_df[~((mod_bot_df < (Q1 - 1.5 * IQR)) |(mod_bot_df > (Q3 + 1.5 * IQR))).any(axis=1)]
mod_bot_df_out.shape
```

## 3.	Perform linear regression on the dataset to see if you can predict water temperature based on salinity. 

```python
linreg_bot_df = mod_bot_df_out[['Salnty', 'T_degC']]
linreg_bot_df
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression

X = linreg_bot_df['Salnty'].values
y = linreg_bot_df['T_degC'].values

X_reshaped = X.reshape(-1, 1)
y_reshaped = y.reshape(-1, 1)

prediction_space = np.linspace(min(X), max(X)).reshape(-1,1)

X_train, X_test, y_train, y_test = train_test_split(X_reshaped, y_reshaped, test_size = 0.5, random_state=47)

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
model.score(X_test, y_test)
```

## 4.	Find your mean squared error and R2 values. Be sure to indicate what these tell you. You can import these from sklearn.metrics . 

```python
#Mean squared error
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

```python
#R2 values
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

## 5.	Create a final plot of the relationship between water temperature and salinity. Be sure to show the values from the dataset in a scatter plot with a trend line that shows the predicted temperature values. 

```python
plt.scatter(X_reshaped, y_reshaped)
plt.plot(X_test, y_pred, color='black', linewidth=3)
plt.xlabel("Salinity")
plt.ylabel("Temperature")
plt.show()
```

### When a linear model doesn't fit the data as well as preferred and the data can be better fit through a more sophisticated model polynomial regression may improve the outcomes.  Polynomial regression is the adding of features to your dataset that increase the values exponentially.  This results in a curved model that may eliminate some underfitting.  It is important not to overfit the data by adding too many degrees, however. 

## 7. Perform polynomial regression on this same dataset. Train and validate your model as you did when you used linear regression. Mean squared error and R2 both apply here. 

```python
linreg_bot_df
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = linreg_bot_df['Salnty'].values
y = linreg_bot_df['T_degC'].values

X_reshaped = X.reshape(-1, 1)
y_reshaped = y.reshape(-1, 1)

scale = StandardScaler()
X_reshaped = scale.fit_transform(X_reshaped)

X_train, X_test, y_train, y_test = train_test_split(X_reshaped, y_reshaped, test_size = 0.2, random_state=42)

poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X_train)
poly.fit(X_train, y_train)

model = LinearRegression()
model.fit(X_poly, y_train)

y_pred = model.predict(poly.fit_transform(X_test))
model.score((poly.fit_transform(X_test)), y_test)
```
